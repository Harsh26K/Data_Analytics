```python
avg_cost_across_country = []
currencies = []
countries = final_df['Country'].unique()

for country in countries:
  sum=0
  for i in range(0, len(final_df)):
    if final_df['Country'][i] == country:
      currency = final_df['Currency'][i]
      sum+=final_df['Average Cost for two'][i]
      j=i+1
    acac = sum/j
  avg_cost_across_country.append(acac)
  currencies.append(currency)
  
d= {'Country':countries,'AvgCostAcrossCountry':avg_cost_across_country,'Currency':currencies}
  
Avg_Cost_for_two_Across_Country=pd.DataFrame(data=d)
```
